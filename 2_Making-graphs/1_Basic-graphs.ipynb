# Basic plotting
## Load in your data

````python
import pandas as pd
import numpy as np
spreadsheet_file = 'example_data.xlsx'
df = pd.read_excel(spreadsheet_file)
df.head(3)
````

## Create a plot

We will be using a program called seaborn. It creates nice looking graphs without much effort from the user.

It uses dataframes as the input, which means that sometimes your data has to be organised in a specific way for it to work.

It can create heaps of different graphs https://seaborn.pydata.org/examples/index.html

````python
import seaborn as sns
````

The fundamental building blocks of any seaborn plot are inputting an x variable, a y variable and the dataframe to draw these from.

Below I input x='CLIN_STATUS',y='COG_MEM',data=df. Note that the x and y labels are __exactly the same as the columns in the df.__

````python
sns.barplot(x='CLIN_STATUS',y='AGE',data=df)
````

We can plot this data in many different ways by using the same input but a different seaborn command. 

See here for a box plot:

````python
sns.boxplot(x='CLIN_STATUS',y='AGE',data=df)
````

Or a scatterplot:

````python
sns.lmplot(x='CLIN_STATUS',y='AGE',data=df)
````

Notice that the commands inside the brackets stay exactly the same? 
```python
(x='CLIN_STATUS',y='AGE',data=df)
```
This makes it easy to experiment with different plotting styles.

The bit that changes is the seaborn command, e.g.,
```python
sns.barplot
sns.boxplot
sns.lmplot
```

## Adding a third 'z' colour dimension

Often in plots you will want to add another dimension beyond the x and y axes. This might be clinical group, where you plot each group in a different colour.

Seaborn implements this using the 'hue' argument. See below for an example.

````python
sns.barplot(x='CLIN_STATUS',y='AGE',hue='SEX',data=df)
````

## Adding a fourth dimension!

You can split your plots up in another way - by plotting more plots! 

Say you wanted to recreate the above graph - age by clinical Status in both sexes - but looking at patterns across genetic status. You could manually create a new plot for each status by altering the dataframe. Or you can use seaborns 'meta' plot commands which are built to perform many plots at once.

````python
sns.catplot(x='CLIN_STATUS',y='AGE',hue='SEX',col='GENETIC_STATUS',kind='bar',data=df)
````

There are two extra arguments here: 
- 'col' (i.e., column) where you specific what variable should be plotting across the columns - in this case, 'GENETIC_STATUS'.
- 'kind' which is the type of plot you would like. In this case I specified bar (as in barplot).

Here is an example using box and then strip.

````python
sns.catplot(x='CLIN_STATUS',y='AGE',hue='SEX',col='GENETIC_STATUS',kind='box',data=df)
````

````python
sns.catplot(x='CLIN_STATUS',y='AGE',hue='SEX',col='GENETIC_STATUS',kind='strip',data=df)
````
